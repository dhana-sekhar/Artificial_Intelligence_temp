# KNN - Vsiualization 

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
from sklearn import datasets, neighbors

def knn_Nearbor(data, n_neighbor):

  x = data[:, :2]
  y = data[:, 2]
  h = .02
  x_min , x_max = data[:, :1].min(), data[:, :1].max()
  y_min, y_max = data[:, 1].min(), data[:, 1].max()
  xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
  clf = neighbors.KNeighborsClassifier(n_neighbor)
  clf.fit(x,y)
  z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
  z = z.reshape(xx.shape)
  cmap_light = ListedColormap(['#FFAAAA',  '#AAAAFF'])
  cmap_bold = ListedColormap(['#FF0000', '#0000FF'])
  plt.figure()
  plt.pcolormesh(xx, yy, z, cmap= cmap_light)
  plt.scatter(x[:, 0], x[:, 1], c=y, cmap=cmap_bold)
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
  plt.title('K value = '+str(n_neighbor)) 

```

# Creating Cross-Validation

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```
data = pd.read_csv("/content/drive/My Drive/dataset_aaic/demo_data/3.concertriccir2.csv", names = ['x', 'y', 'class'])
```

```
x = np.array(data.iloc[:, 0:2])
y = np.array(data['class'])
```

```
x.shape
```

```
y.shape
```

```
# this is first train test split
x_train1, x_test, y_train1, y_test = train_test_split(x, y, test_size = 0.3)
```

```
# this is going to be the second train test split
x_train, x_cv, y_train, y_cv = train_test_split(x_train1, y_train1, test_size = 0.3)
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```
cv_acc = []
for i in range(1, 50):
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(x_train, y_train)
  pred = knn.predict(x_cv)
  acc = accuracy_score(y_cv, pred, normalize = True)
  print("The accuracy of Cross-Validation of {}NN is : {}".format(i, acc*100))
  cv_acc.append((i, acc*100))
cv_acc = np.array(cv_acc)
```

```
test_acc = []
for i in range(1,50):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train, y_train)
  pred = knn.predict(x_test)
  acc = accuracy_score(y_test, pred, normalize = True)
  print("*** The Accuracy for Test data when {}NN is : {}".format(i, acc*100))
  test_acc.append((i, acc*100))
```

```
test_acc = np.array(test_acc)
```

```
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```
data = pd.read_csv("/content/drive/My Drive/dataset_aaic/demo_data/3.concertriccir2.csv", names = ['x', 'y', 'class'])
x = np.array(data.iloc[:, 0:2])
y = np.array(data['class'])
x_tr1, x_test, y_tr1, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
x_train, x_cv, y_train, y_cv = train_test_split(x_tr1, y_tr1, test_size = 0.3)
for i in range(1,50):
  knn = KNeighborsClassifier(n_neighbors=i, weights = "distance", n_jobs = -1)
  knn.fit(x_train, y_train)
  pred = knn.predict(x_cv)
  acc = accuracy_score(y_cv, pred, normalize=True)
  print("The CV score for {} NN is : {}". format(i, acc*100))

for i in range(1,50):
  knn = KNeighborsClassifier(n_neighbors = i, weights='distance', n_jobs=-1)
  knn.fit(x_train, y_train)
  pred = knn.predict(x_test)
  acc = accuracy_score(y_test, pred, normalize=True)
  print("The acutal Accuracy for {} NN is : {}".format(i, acc*100))
```

```
from sklearn.model_selection import cross_val_score
```

```
data = pd.read_csv("/content/drive/My Drive/dataset_aaic/demo_data/3.concertriccir2.csv", names = ['x', 'y', 'class'])
x = np.array(data.iloc[:, 0:2])
y = np.array(data['class'])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
cv_score = []
for i in range(1,50):
  knn = KNeighborsClassifier(n_neighbors=i, weights='distance', n_jobs=-1)
  score = cross_val_score(knn, x_train, y_train, cv = 10, n_jobs=-1)
  cv_score.append(score.mean())
```
