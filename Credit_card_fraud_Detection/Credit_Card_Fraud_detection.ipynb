# Data Creation

```
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv("/content/drive/My Drive/DATASETS/creditcard.csv")
train, test = train_test_split(data, test_size=0.2)
train.head()
```

# Setup Data

```
from pycaret.classification import *

clf = setup(data = train, target = 'Class')
```

# Compare Models

```
compare_models()
```

# Create Model

```
mod = 'xgboost'
model = create_model(mod)
```

# Tune Model

```
model = tune_model(mod)
```

# Predict Model

```
test_data = test.drop('Class', axis=1)
test_data.head()
```

```
out_put = predict_model(model, data = test_data)
```

```
out_put.head()
```

```
import numpy as np
from sklearn.metrics import accuracy_score
acc = accuracy_score(np.array(test['Class']),out_put['Label'])
```

```
print("The accuracy of the the model is : {}".format(acc*100))
```
